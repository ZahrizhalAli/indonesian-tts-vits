# **Setup Config**

1. Pindahkan model dari "/root/.local/share/tts/tts_models--multilingual--multi-dataset--your_tts" pada direktori GDrive anda "/content/drive/MyDrive/tts_models--multilingual--multi-dataset--your_tts"

2. Kemudian, hapus file speakers.json dan lakukan compute embedding kembali pada dataset baru.

3. Setup config.json sesuai kebutuhan dataset anda :
    - output_path --> /content/drive/MyDrive/yourtts/
    - run_name --> Bebas
    - audio {sample_rate : 22050} -->Depend on dataset
    -use_phonemes : false
    - phoneme_language :id
    - text_cleaners
    - dataset {formatter : "ljspeech", path : to/directory/dataset/, meta_file_train : filename_script.txt, language : "id"}
    - test_sentences
    - d_vector_file : /content/drive/MyDrive/yourtts/speakers.json --> List
    - speaker_encoder_config_path: config_se.json
    - speaker_encoder_model_path : model_se.pth

4. Setup config_se.json sesuai kebutuhan dataset anda :
    - output_path --> /content/drive/MyDrive/yourtts/
    - audio {sample_rate : 22050}

5. Anda juga dapat merubah hyper-parameter (epoch, save_step, batch_size, dll) pada config.json sesuai kebutuhan

6. Jalankan code dibawah untuk menghitung ulang embedding pada dataset.
